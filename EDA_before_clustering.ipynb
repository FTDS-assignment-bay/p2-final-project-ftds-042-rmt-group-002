# **4. Exploratory Data Analysis (EDA) - Before Clustering**
---
Kita akan melanjutkan ke tahap *Exploratory Data Analysis (EDA)* sebelum dilakukan clustering pelanggan churn. Pada bagian ini, dilakukan eksplorasi untuk memahami data lebih dalam dan menemukan pola terkait pelanggan *churn*.

### **4.1. Distribusi Umur Pelanggan Churn**
 
Pada bagian ini, kita akan melihat distribusi umur pelanggan `churn`.

```python
# Membuat barplot
plt.figure(figsize=(8,5))
sns.histplot(df_churn['age'], bins=20, kde=True, color='skyblue')
plt.title('Distribusi Umur Pelanggan')
plt.xlabel('Umur')
plt.ylabel('Jumlah Pelanggan')
plt.tight_layout()
plt.show()

# Menmbuat variable untuk menghitung jumlah dan persentase umur pelanggan
age_counts = df_churn['age'].value_counts().sort_index()
age_percent = (age_counts / age_counts.sum() * 100).round(2)

# Gabungkan jumlah dan persentase ke dalam DataFrame
age_summary = pd.DataFrame({
    'Jumlah Pelanggan': age_counts,
    'Persentase (%)': age_percent
})

# Tampilkan hasil
print("Umur dan Jumlah Pelanggan Churn:")
print(age_summary)
```

Insight:
1. Mayoritas pelanggan yang telah churn berada dalam usia produktif. Sebesar 47.76% pelanggan churn berusia 30 tahun dan 35.20% lainnya berusia 25 tahun. Berdasarkan hasil ini, kemungkinan para pelanggan tersebut berhenti berlangganan untuk menggunakan produk dari kompetitor dengan beberapa alasan tertentu.
2. Terdapat 74 orang atau sekitar 16,59% pelanggan churn yang berusia 45 tahun. Seperti halnya pelanggan churn di usia sebelumnya, kemungkinan pelanggan churn pada usia ini berhenti berlangganan karena menggunakan produk dari kompetitor.
3. Hanya terdapat 2 pelanggan churn yang berusia 55 tahun. Hal ini mungkin disebabkan karena mereka berhenti menggunakan gadget secara rutin, atau mungkin juga menggunakan produk dari kompetitor.

### **4.2. Distribusi Lama Berlangganan (`subscription_length`)**



Selanjutnya, kita akan melihat bagaimana distribusi lama berlangganan pelanggan sebelum `churn`.

```python
# Membuat Plot distribusi subscription_length
plt.figure(figsize=(8, 5))
sns.histplot(df_churn['subscription_length'], bins=40, kde=True, color='salmon')
plt.title('Distribusi Lama Berlangganan - Pelanggan Churn')
plt.xlabel('Lama Berlangganan (bulan)')
plt.ylabel('Jumlah Pelanggan')
plt.tight_layout()
plt.show()

# Membuat variable untuk menghitung frekuensi setiap nilai subscription_length
subscription_counts = df_churn['subscription_length'].value_counts().sort_values(ascending=False)

# Mengambil 5 nilai jumlah pelanggan teratas
top_5_subscription_freq = subscription_counts.sort_values(ascending=False).head(5)
top_5_subscription_freq_df = top_5_subscription_freq.reset_index()
top_5_subscription_freq_df.columns = ['subscription_length', 'jumlah_pelanggan']

# Mengambil 5 nilai jumlah pelanggan terendah
least_5_subscription_freq = subscription_counts.sort_values(ascending=True).head(5)
least_5_subscription_freq_df = least_5_subscription_freq.reset_index()
least_5_subscription_freq_df.columns = ['subscription_length', 'jumlah_pelanggan']

# Input hasil ke DataFrame 
top_5_subscription_freq_df = top_5_subscription_freq.reset_index()
top_5_subscription_freq_df.columns = ['subscription_length', 'jumlah_pelanggan']

# Menampilkan hasil
print("Top 5 Jumlah Pelanggan Berdasarkan Subscription Length:")
print(top_5_subscription_freq_df)
print("Least 5 Jumlah Pelanggan Berdasarkan Subscription Length:")
print(least_5_subscription_freq_df)
```

Insight:
1. Pelanggan churn terbanyak rata-rata telah berlangganan selama kurang lebih 3 tahun.
2. Pelanggan churn dengan jumlah paling sedikit memiliki masa berlangganan yang bervariasi. Terdapat masing-masing seorang pelanggan yang telah berlangganan sekitar 1,5-2 tahun serta seorang pelanggan yang telah berlangganan hampir selama 4 tahun. Pelanggan-pelanggan ini mungkin dapat dianggap sebagai data outlier di analisis selanjutnya.
3. Secara keseluruhan, pelanggan churn telah berlangganan lebih dari 1 tahun. Pelanggan churn mungkin tidak disebabkan oleh ketidakpuasan terhadap pelayanan saat awal mereka mulai berlangganan, sehingga perlu dilakukan analisis lebih lanjut dengan fitur lainnya untuk mengetahui faktor lain yang mempengaruhi keputusan untuk berhenti berlangganan.

### **4.3. Distribusi Fitur Complain (`complains`)**

Pada bagian ini, kita akan melihat jumlah pelanggan `churn` berdasarkan status pengajuan keluhan.

```python
# Membuat barplot
plt.figure(figsize=(6, 4))
sns.countplot(data=df_churn, x='complains', palette='pastel')
plt.title('Distribusi Keluhan pada Pelanggan Churn')
plt.xlabel('Pengajuan Keluhan')
plt.ylabel('Jumlah Pelanggan')
plt.tight_layout()
plt.show()

# Membuat variable untuk menghitung jumlah pelanggan churn berdasarkan status keluhan
complaint_counts = df_churn['complains'].value_counts().reset_index()
complaint_counts.columns = ['complains', 'jumlah_pelanggan']

# Menampilkan hasil
print("Jumlah Pelanggan Churn Berdasarkan Keluhan:")
print(complaint_counts)
```

### **4.4. Distribusi Aktivitas Penggunaan Produk**



Pada bagian ini, kita akan distribusi pelanggan churn berdasarkan aktivitas yang dilakukan melalui fitur `seconds_of_use`, `frequency_of_use` dan `frequency_of_sms`.

```python
# Mendefinisikan ukuran subplot
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
sns.set(style="whitegrid")

# Membuat subplot 1, yaitu distribusi seconds_of_use
sns.histplot(df_churn['seconds_of_use'], bins=30, kde=True, color='skyblue', ax=axes[0])
axes[0].set_title('Distribusi Waktu Penggunaan')
axes[0].set_xlabel('Total Detik Penggunaan Panggilan')
axes[0].set_ylabel('Jumlah Pelanggan')

# Membuat subplot 2, yaitu distribusi frequency_of_use
sns.histplot(df_churn['frequency_of_use'], bins=30, kde=True, color='lightgreen', ax=axes[1])
axes[1].set_title('Distribusi Frekuensi Penggunaan')
axes[1].set_xlabel('Frekuensi Penggunaan')
axes[1].set_ylabel('Jumlah Pelanggan')

# Membuat subplot 3, yaitu distribusi frequency_of_sms
sns.histplot(df_churn['frequency_of_sms'], bins=30, kde=True, color='salmon', ax=axes[2])
axes[2].set_title('Distribusi Frekuensi SMS')
axes[2].set_xlabel('Frekuensi SMS')
axes[2].set_ylabel('Jumlah Pelanggan')

# Menampilkan subplot
plt.tight_layout()
plt.show()

# Top 5 pelanggan churn berdasarkan Seconds of Use
top5_seconds = df_churn['seconds_of_use'].value_counts().sort_values(ascending=False).head(5).reset_index()
top5_seconds.columns = ['seconds_of_use', 'jumlah_pelanggan']
print("Top 5 Seconds of Use:")
print(top5_seconds)

# Top 5 pelanggan churn berdasarkan Frequency of Use
top5_freq_use = df_churn['frequency_of_use'].value_counts().sort_values(ascending=False).head(5).reset_index()
top5_freq_use.columns = ['frequency_of_use', 'jumlah_pelanggan']
print("\nTop 5 Frequency of Use:")
print(top5_freq_use)

# Top 5 pelanggan churn berdasarkan Frequency of SMS
top5_sms = df_churn['frequency_of_sms'].value_counts().sort_values(ascending=False).head(5).reset_index()
top5_sms.columns = ['frequency_of_sms', 'jumlah_pelanggan']
print("\nTop 5 Frequency of SMS:")
print(top5_sms)
```

Insight:
1. Berdasarkan durasi penggunaan panggilannya, sebanyak 56 pelanggan berhenti berlangganan tanpa menggunakan produk langganan yang telah dibeli. Sebagian dari pelanggan memutuskan untuk berhenti pelangganan setelah menggunakan panggilan minimal selama 5 menit.
2. Berdasarkan frekuensi penggunaannya, sebanyak 56 pelanggan berhenti berlangganan tanpa pernah menggunakan fitur panggilan. 
3. Berdasarkan frekuensi penggunaan SMS, sebanyak 109 pelanggan berhenti berlangganan tanpa pernah menggunakan fitur SMS.
4. Dari distribusi data secara keseluruhan, mayoritas pelanggan churn adalah pelanggan pasif yang tidak pernah memakai layanan produk. Perlu ditelusuri lebih lanjut mengapa mereka cenderung tidak menggunakan layanan. Sementara itu, untuk pelanggan lain yang telah memakai layanan produk dan akhirnya memilih untuk berhenti berlangganan, kemungkinan dapat disebabkan oleh adanya ketidakpuasan saat penggunaan layanan tertentu.


### **4.5. Korelasi Antar Fitur pada Pelanggan `churn`**



Pada bagian ini, kita akan melihat hubungan dari masing-masing fitur pada pelanggan `churn`.

```python
# Membuat heatmap
plt.figure(figsize=(14,10))
corr = df_churn.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Heatmap Korelasi Antar Fitur')
plt.show()
```

Insight:
1. Korelasi tertinggi terdapat pada fitur `seconds_of_use` dan `frequency_of_use`, yaitu sebesar 0.89. Hal ini menunjukkan bahwa pelanggan churn adalah pelanggan aktif yang sering beraktivitas menggunakan produk layanan dari perusahaan.
2. Terdapat korelasi yang cukup tinggi pada penggunaan sms `frequency_of_sms`dan `customer_value`, yaitu sebesar 0.83. Hal ini menunjukkan bahwa pelanggan churn aktif memakai layanan sms dan memiliki nilai `customer_value` yang cukup baik.
3. Terdapat korelasi negatif yang cukup tinggi pada `status` keaktifan pelanggan dengan `charge_amounts`, yaitu dengan nilai -0.64. Hal ini menunjukkan bahwa tidak ada hubungan antara keaktifan pelanggan churn dengan besaran biaya tambahan yang telah dikeluarkan untuk berlangganan.
4. Terdapat korelasi negatif pada fitur `charge_amounts` dengan lama berlangganan (`subscription_length`), yaitu dengan nilai -0.44. Hal ini menunjukkan bahwa besaran biaya tambahan saat berlangganan pada pelanggan churn tidak memiliki pengaruh yang kuat dengan lama berlangganan sebelum berhenti.
5. Secara keseluruhan, pelanggan yang telah berhenti berlangganan adalah pelanggan aktif yang telah menggunakan beberapa layanan yang tersedia dan memiliki nilai `customer_value` yang cukup baik. Namun, kemungkinan terdapat adanya ketidakpuasan pada layanan selama memiliki paket berlangganan, baik pada fitur produk yang ditawarkan, ataupun layanan yang dilakukan langsung oleh pihak perusahaan (semisal melalui customer service).
