```python
# Membuat variable untuk menghitung metrik masing-masing grafik
churn_by_complains = data.groupby('complains')['churn'].mean()
churn_by_call_failure = data.groupby(pd.cut(data['call_failure'], bins=5))['churn'].mean()
churn_by_customer_value = data.groupby(pd.cut(data['customer_value'], bins=5,))['churn'].mean()

# Membuat ukuran subplot
fig, axs = plt.subplots(1, 2, figsize=(20, 6))

# Membuat subplot 1, yaitu Churn berdasarkan Complaints
axs[0].pie(churn_by_complains, labels=['No Complaints', 'Complaints'], autopct='%1.1f%%',
           startangle=90, colors=['#66b3ff', '#ff9999'])
axs[0].set_title('Churn Rate berdasarkan Keluhan')
axs[0].axis('equal')

# Membuat subplot 2, yaitu Churn berdasarkan Call Failures
call_failure_bins = pd.cut(data['call_failure'], bins=[0, 5, 10, 15, 20, data['call_failure'].max()])
churn_by_call_failure = data.groupby(call_failure_bins)['churn'].mean()
axs[1].bar(churn_by_call_failure.index.astype(str), churn_by_call_failure.values, color='skyblue')
axs[1].set_title('Churn Rate berdasarkan Call Failures')
axs[1].set_xlabel('Call Failure Bins')
axs[1].set_ylabel('Churn Rate')

plt.tight_layout()
plt.show()
```

### **5.6. Outlier Handling**

```python
# Kolom numerik yang akan dicek skewness
num_cols = [
    'call_failure',
    'subscription_length',
    'seconds_of_use',
    'frequency_of_use',
    'frequency_of_sms',
    'distinct_called_numbers',
    'age',
    'customer_value'
]

# Cek skewness
listItem = []
for col in num_cols:
    skew_value = X_train[col].skew()
    distribusi = 'normal' if -0.5 <= skew_value <= 0.5 else 'skewed'
    listItem.append([col, round(skew_value, 3), distribusi])

skewness = pd.DataFrame(listItem, columns=['kolom', 'nilai skewness', 'distribusi'])
skewness
```

Untuk memahami karakteristik distribusi dari masing-masing fitur numerik, dilakukan perhitungan nilai skewness. Skewness mengukur derajat kemiringan distribusi data; nilai mendekati 0 mengindikasikan distribusi yang simetris (normal), sedangkan nilai positif atau negatif yang besar menunjukkan distribusi yang miring ke kanan atau kiri.

Berikut hasil perhitungan skewness pada fitur numerik:
* Seluruh fitur numerik menunjukkan distribusi yang **skewed**, baik ke kanan (positif) maupun ke kiri (negatif). Hal ini penting untuk diperhatikan karena:
    - Algoritma machine learning berbasis jarak (seperti KNN, SVM) dan berbasis linearitas (seperti Logistic Regression) cenderung sensitif terhadap distribusi yang tidak normal.
    - Skewness yang tinggi dapat mempengaruhi stabilitas model dan interpretasi fitur.


**Handling Skewness: Winsorization**

Seluruh fitur numerik dalam dataset menunjukkan adanya skewness, baik ke kanan maupun ke kiri. Oleh karena itu, dilakukan penanganan skewness dengan menggunakan teknik **Winsorization** berbasis Interquartile Range (IQR). Winsorization merupakan metode capping outlier yang menjaga sebagian besar nilai data tetap dalam skala aslinya, namun membatasi pengaruh outlier ekstrem.

Parameter Winsorizer yang digunakan:

- Metode: IQR
- Tail: Both (kanan dan kiri)
- Fold: 1.5 (batas IQR default)


```python
# List kolom numerik hasil analisis skewness
num_cols = [
    'call_failure',
    'subscription_length',
    'seconds_of_use',
    'frequency_of_use',
    'frequency_of_sms',
    'distinct_called_numbers',
    'age',
    'customer_value'
]

# Membuat objek winsorizer
# Di sini saya pakai method 'iqr' karena cukup aman untuk skewness
winsor = Winsorizer(capping_method='iqr',
                    tail='both',  # bisa disesuaikan 'right' jika hanya ingin handle upper outlier
                    fold=1.5,
                    variables=num_cols)

# Fit transformer pada data latih dan transform data latih
X_train_capped = winsor.fit_transform(X_train)
# Transform data uji menggunakan parameter yang sama (dipelajari dari data latih)
X_test_capped = winsor.transform(X_test)
```

```python
# Membuat visualisasi histogram dan boxplot
def plots(df, variables):
    plt.figure(figsize=(16, 4))

    # Menampilkan Histogram
    plt.subplot(1, 2, 1)
    sns.histplot(df[variables], bins=30, color='crimson')
    plt.title('Histogram')

    # Menampilkan Boxplot
    plt.subplot(1, 2, 2)
    sns.boxplot(y=df[variables], color='crimson')
    plt.title('Boxplot')

    # Menampilkan visualisasi
    plt.show()

# Melihat distribusi data sebelum dan setelah dilakukan capping
for col in num_cols:
  print(f'{col} - Sebelum dilakukan Capping')
  plots(X_train, col)

  print(f'\n{col} - Sesudah dilakukan Capping')
  plots(X_train_capped, col)
```

Visualisasi boxplot mengonfirmasi bahwa penanganan nilai ekstrem melalui teknik capping berhasil menangani outlier yang sebelumnya ada pada kolom-kolom yang dianalisis.

### **5.7. Feature Selection**

#### **5.7.1. Menguji Korelasi antar Fitur dengan Target**

```python
# Menggabungkan fitur dan target data latih untuk perhitungan Phik matrix
train_data = pd.concat([X_train_capped, y_train], axis=1)

# Membuat fungsi untuk menghitung Phi-K correlation
def compute_phik_correlation(dataframe, columns, target):

    # Mengambil subset dari DataFrame
    subset = dataframe[columns]
    # Menghitung matriks Phi-K
    correlation_matrix = subset.phik_matrix()
    # Mengembalikan kolom target dari matriks
    return correlation_matrix[target]

# Definisikan list kolom untuk setiap subset
columns = ['call_failure', 'complains', 'subscription_length', 'charge_amount',
       'seconds_of_use', 'frequency_of_use', 'frequency_of_sms',
       'distinct_called_numbers', 'age_group', 'tariff_plan', 'status', 'age',
       'customer_value', 'churn']
target = ['churn']

# Compute Phi-K correlation for each set of columns and print
correlation = compute_phik_correlation(train_data, columns, target)

# Menampilkan hasil
print(correlation)
```

Analisis korelasi dilakukan dengan menggunakan Phi-K correlation karena mampu mengukur hubungan antara fitur numerik dan kategorikal, termasuk hubungan non-linear.

Fitur `complains` dan `status` memiliki pengaruh paling kuat terhadap kemungkinan churn.

Memasukkan fitur dengan relevansi sangat rendah seperti ini berpotensi menambah noise tanpa memberi manfaat berarti bagi model prediksi.

### **5.9. Column Transformer**

Selanjutnya, saya menyiapkan alat untuk pra-pemrosesan, yaitu `ColumnTransformer` yang saya beri nama `preprocessor`.
* Pada kolom numerik (`numerical_cols_selected`) akan diterapkan `StandardScaler` untuk mengubah skala nilainya (rata-rata 0, standar deviasi 1).
* Pada kolom kategorikal ordinal (`ordinal_categorical_cols` seperti `charge_amount`, `age_group`)akan diterapkan `OrdinalEncoder`.
* Pada kolom kategorikal nominal (`categorical_cols_selected` seperti `complains`, `tariff_plan` dll)akan diterapkan `OneHotEncoder`.

```python
preprocessor= ColumnTransformer(
    transformers=[
        ('scaler', StandardScaler(), numerical_cols_selected),
        ('ord_encode', OrdinalEncoder( handle_unknown='use_encoded_value', 
                                      unknown_value=-1), # Mengatasi unknown di test
                                      ordinal_cols_selected),
        ('cat_ohe', OneHotEncoder(handle_unknown='ignore', drop='first', sparse_output=False), 
                          categorical_cols_selected)
    ])
```

    * **Tujuan Tuning**: Mencari kombinasi hyperparameter (`n_estimators`, `max_depth`, `learning_rate`, dll.) untuk memaksimalkan F1-Score dan Recall secara bersamaan.
