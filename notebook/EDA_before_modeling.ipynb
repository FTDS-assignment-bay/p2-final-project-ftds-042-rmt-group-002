### **4.3. Distribusi Fitur Kategorikal (`tariff_plan` dan `status`)**

Analisis ini akan menunjukkan jumlah pelanggan berdasarkan sistem pembayaran langganan (`tariff_plan`) dan `status` aktif pelanggan.

```python
# Mendefinisikan ukuran subplot
fig, axs = plt.subplots(1, 2, figsize=(12,5))

# Membuat subplot 1, yaitu distribusi tariff_plan
sns.countplot(x='tariff_plan', data=data, ax=axs[0], palette='Set1')
axs[0].set_title('Distribusi Tariff Plan')
axs[0].set_xticklabels(['Prabayar', 'Pascabayar'])

# Membuat subplot 2, yaitu distribusi status aktif pelanggan
sns.countplot(x='status', data=data, ax=axs[1], palette='Set2')
axs[1].set_title('Status Aktif Pelanggan')
axs[1].set_xticklabels(['Aktif', 'Non-aktif'])

# Menampilkan subplot
plt.tight_layout()
plt.show()

# Membuat mapping label
tariff_labels = {1: 'Prabayar', 2: 'Pascabayar'}
status_labels = {1: 'Aktif', 2: 'Non-aktif'}

# Menampilkan Output Tariff Plan
tariff_counts = data['tariff_plan'].value_counts()
tariff_percent = data['tariff_plan'].value_counts(normalize=True) * 100
print("Tariff Plan:")
for label, count in tariff_counts.items():
    label_name = tariff_labels.get(label, label)
    print(f"  {label_name} = {count} pelanggan ({tariff_percent[label]:.2f}%)")

# Menampilkan Output Status
status_counts = data['status'].value_counts()
status_percent = data['status'].value_counts(normalize=True) * 100
print("\nStatus Pelanggan:")
for label, count in status_counts.items():
    label_name = status_labels.get(label, label)
    print(f"  {label_name} = {count} pelanggan ({status_percent[label]:.2f}%)")
```

Insight:
1. Mayoritas pelanggan, yaitu 2.621 orang atau sekitar 92% pelanggan, berlangganan paket prabayar. Sedangkan sebagian kecil sisanya berlangganan paket pascabayar. Hal ini menunjukkan jika pelanggan memiliki preferensi untuk memlakukan pembelian paket langganan berdasarkan biaya / kuota yang dibutuhkan.
2. Mayoritas pelanggan, yaitu 2.166 orang atau sekitar 76% pelanggan, berstatus aktif. Sedangkan sisanya termasuk tidak aktif.
3. Dari data tersebut, perlu dilihat lebih detail kembali mengenai status pelanggan aktif dan non-aktif, apakah sebagian besar adalah pelanggan prabayar atau sebaliknya.


### **4.4. Korelasi `complains` dan `call_failures` dengan `churn` Rate**



Pada bagian ini, kita akan melihat korelasi pelanggan yang melakukan `complains` dan `call_failures` terhadap keputusan `churn`, serta nilai pelanggan yang `churn`.

```python
# Membuat variable untuk menghitung metrik masing-masing grafik
churn_by_complains = data.groupby('complains')['churn'].mean()
churn_by_call_failure = data.groupby(pd.cut(data['call_failure'], bins=5))['churn'].mean()
churn_by_customer_value = data.groupby(pd.cut(data['customer_value'], bins=5,))['churn'].mean()

# Membuat ukuran subplot
fig, axs = plt.subplots(1, 2, figsize=(20, 6))

# Membuat subplot 1, yaitu Churn berdasarkan Complaints
axs[0].pie(churn_by_complains, labels=['No Complaints', 'Complaints'], autopct='%1.1f%%',
           startangle=90, colors=['#66b3ff', '#ff9999'])
axs[0].set_title('Churn Rate berdasarkan Keluhan')
axs[0].axis('equal')

# Membuat subplot 2, yaitu Churn berdasarkan Call Failures
call_failure_bins = pd.cut(data['call_failure'], bins=[0, 5, 10, 15, 20, data['call_failure'].max()])
churn_by_call_failure = data.groupby(call_failure_bins)['churn'].mean()
axs[1].bar(churn_by_call_failure.index.astype(str), churn_by_call_failure.values, color='skyblue')
axs[1].set_title('Churn Rate berdasarkan Call Failures')
axs[1].set_xlabel('Call Failure Bins')
axs[1].set_ylabel('Churn Rate')

plt.tight_layout()
plt.show()
```

Dari distribusi data di atas, kita mendapatkan beberapa informasi, yaitu:
1. Mayoritas pelanggan yang telah churn pernah mengajukan keluhan, yaitu sekitar 89.4% dari total pelanggan churn. Hal ini menunjukkan bahwa pelanggan yang pernah atau sering menyampaikan keluhan, cenderung untuk berhenti berlangganan di suatu waktu tertentu.
2. Sekitar 21% pelanggan churn setelah mengalami call failure lebih dari 20 kali, sedangkan sisanya juga berhenti berlangganan walaupun tidak terlalu sering mengalamki call failure.
3. Dengan tingginya tingkat complain pada pelanggan churn serta banyaknya call failure yang dialami mereka saat masih aktif, terlihat bahwa pelanggan yang telah berhenti berlangganan sebagian besar mengalami kendala dalam menggunakan paket langganannya dan telah menyampaikan keluhannya kepada perusahaan.
