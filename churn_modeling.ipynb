### **5.6. Outlier Handling**

```python
# Kolom numerik yang akan dicek skewness
num_cols = [
    'call_failure',
    'subscription_length',
    'seconds_of_use',
    'frequency_of_use',
    'frequency_of_sms',
    'distinct_called_numbers',
    'age',
    'customer_value'
]

# Cek skewness
listItem = []
for col in num_cols:
    skew_value = X_train[col].skew()
    distribusi = 'normal' if -0.5 <= skew_value <= 0.5 else 'skewed'
    listItem.append([col, round(skew_value, 3), distribusi])

skewness = pd.DataFrame(listItem, columns=['kolom', 'nilai skewness', 'distribusi'])
skewness
```

Untuk memahami karakteristik distribusi dari masing-masing fitur numerik, dilakukan perhitungan nilai skewness. Skewness mengukur derajat kemiringan distribusi data; nilai mendekati 0 mengindikasikan distribusi yang simetris (normal), sedangkan nilai positif atau negatif yang besar menunjukkan distribusi yang miring ke kanan atau kiri.

Berikut hasil perhitungan skewness pada fitur numerik:
* Seluruh fitur numerik menunjukkan distribusi yang **skewed**, baik ke kanan (positif) maupun ke kiri (negatif). Hal ini penting untuk diperhatikan karena:
    - Algoritma machine learning berbasis jarak (seperti KNN, SVM) dan berbasis linearitas (seperti Logistic Regression) cenderung sensitif terhadap distribusi yang tidak normal.
    - Skewness yang tinggi dapat mempengaruhi stabilitas model dan interpretasi fitur.


**Handling Skewness: Winsorization**

Seluruh fitur numerik dalam dataset menunjukkan adanya skewness, baik ke kanan maupun ke kiri. Oleh karena itu, dilakukan penanganan skewness dengan menggunakan teknik **Winsorization** berbasis Interquartile Range (IQR). Winsorization merupakan metode capping outlier yang menjaga sebagian besar nilai data tetap dalam skala aslinya, namun membatasi pengaruh outlier ekstrem.

Parameter Winsorizer yang digunakan:

- Metode: IQR
- Tail: Both (kanan dan kiri)
- Fold: 1.5 (batas IQR default)


```python
# List kolom numerik hasil analisis skewness
num_cols = [
    'call_failure',
    'subscription_length',
    'seconds_of_use',
    'frequency_of_use',
    'frequency_of_sms',
    'distinct_called_numbers',
    'age',
    'customer_value'
]

# Membuat objek winsorizer
# Di sini saya pakai method 'iqr' karena cukup aman untuk skewness
winsor = Winsorizer(capping_method='iqr',
                    tail='both',  # bisa disesuaikan 'right' jika hanya ingin handle upper outlier
                    fold=1.5,
                    variables=num_cols)

# Fit transformer pada data latih dan transform data latih
X_train_capped = winsor.fit_transform(X_train)
# Transform data uji menggunakan parameter yang sama (dipelajari dari data latih)
X_test_capped = winsor.transform(X_test)
```

```python
# Membuat visualisasi histogram dan boxplot
def plots(df, variables):
    plt.figure(figsize=(16, 4))

    # Menampilkan Histogram
    plt.subplot(1, 2, 1)
    sns.histplot(df[variables], bins=30, color='crimson')
    plt.title('Histogram')

    # Menampilkan Boxplot
    plt.subplot(1, 2, 2)
    sns.boxplot(y=df[variables], color='crimson')
    plt.title('Boxplot')

    # Menampilkan visualisasi
    plt.show()

# Melihat distribusi data sebelum dan setelah dilakukan capping
for col in num_cols:
  print(f'{col} - Sebelum dilakukan Capping')
  plots(X_train, col)

  print(f'\n{col} - Sesudah dilakukan Capping')
  plots(X_train_capped, col)
```

Visualisasi boxplot mengonfirmasi bahwa penanganan nilai ekstrem melalui teknik capping berhasil menangani outlier yang sebelumnya ada pada kolom-kolom yang dianalisis.

### **5.7. Feature Selection**

#### **5.7.1. Menguji Korelasi antar Fitur dengan Target**

```python
# Menggabungkan fitur dan target data latih untuk perhitungan Phik matrix
train_data = pd.concat([X_train_capped, y_train], axis=1)

# Membuat fungsi untuk menghitung Phi-K correlation
def compute_phik_correlation(dataframe, columns, target):

    # Mengambil subset dari DataFrame
    subset = dataframe[columns]
    # Menghitung matriks Phi-K
    correlation_matrix = subset.phik_matrix()
    # Mengembalikan kolom target dari matriks
    return correlation_matrix[target]

# Definisikan list kolom untuk setiap subset
columns = ['call_failure', 'complains', 'subscription_length', 'charge_amount',
       'seconds_of_use', 'frequency_of_use', 'frequency_of_sms',
       'distinct_called_numbers', 'age_group', 'tariff_plan', 'status', 'age',
       'customer_value', 'churn']
target = ['churn']

# Compute Phi-K correlation for each set of columns and print
correlation = compute_phik_correlation(train_data, columns, target)

# Menampilkan hasil
print(correlation)
```

Analisis korelasi dilakukan dengan menggunakan Phi-K correlation karena mampu mengukur hubungan antara fitur numerik dan kategorikal, termasuk hubungan non-linear.

Fitur `complains` dan `status` memiliki pengaruh paling kuat terhadap kemungkinan churn.



#### **5.7.2. Multikolinearitas**

```python
X_vif = train_data[['call_failure',
    'subscription_length',
    'seconds_of_use',
    'frequency_of_use',
    'frequency_of_sms',
    'distinct_called_numbers',
    'age',
    'customer_value']]

# Menghitung VIF untuk setiap fitur
vif_data = pd.DataFrame()
vif_data["fitur"] = X_vif.columns
vif_data["VIF"] = [variance_inflation_factor(X_vif.values, i) for i in range(X_vif.shape[1])]

print(vif_data)
```

Untuk memastikan tidak terjadi redundansi antar fitur numerik, dilakukan pengukuran Variance Inflation Factor (VIF).
Beberapa fitur numerik memiliki VIF tinggi yang menunjukkan adanya potensi multikolinearitas.

Berdasarkan analisis korelasi Phi-K (untuk menilai relevansi fitur terhadap target churn) serta analisis multikolinearitas menggunakan VIF (Variance Inflation Factor), saya memutuskan untuk mengecualikan beberapa fitur berikut dari pemodelan. Adapun pertimbangannya sebagai berikut:
1. `call_failure` (Phik ~0 dan VIF 3.32):

    Meskipun VIF fitur ini relatif rendah (tidak mengalami multikolinearitas tinggi), namun nilai korelasi Phi-K terhadap target `churn` sangat kecil, yaitu sekitar 0. Artinya, variabel ini hampir tidak memiliki kontribusi informasi dalam menjelaskan variasi terjadinya `churn`. Memasukkan fitur dengan relevansi sangat rendah seperti ini berpotensi menambah noise tanpa memberi manfaat berarti bagi model prediksi.

2. `age` (Phik 0.129 dan VIF 15.95):

    Meskipun usia pelanggan secara umum sering diasumsikan berkorelasi dengan perilaku churn, dalam hal ini sudah tersedia fitur `age_group` yang merupakan representasi dari variabel `age` dalam bentuk pengelompokan kategori umur. Dengan tetap mempertahankan `age_group`, informasi terkait usia pelanggan tetap terwakili dalam model. Selain itu, `age` juga menunjukkan multikolinearitas cukup tinggi (VIF > 10), sehingga penghapusan fitur ini dapat membantu mengurangi redundansi informasi.

3. `frequency_of_sms` (Phik 0.339 dan VIF 14.89):

    Walaupun korelasi Phi-K terhadap target tergolong sedang (0.339), fitur ini menunjukkan multikolinearitas tinggi (VIF > 10), diduga karena adanya overlap informasi dengan fitur penggunaan lainnya seperti `frequency_of_use` dan `seconds_of_use`. Oleh karena itu, untuk menghindari redundansi dan menjaga stabilitas model, fitur ini diputuskan untuk dikeluarkan.


```python
# Daftar kolom yang akan dihapus
columns_to_drop = [
    'call_failure', 
    'age',
    'frequency_of_sms'     
]
# Menghapus kolom dari X_train_capped
X_train_selected = X_train_capped.drop(columns=columns_to_drop)
# Menghapus kolom dari X_test_capped
X_test_selected = X_test_capped.drop(columns=columns_to_drop)

# Menampilkan 5 data pertama
X_train_selected.head()
```

### **5.8. Split between Numerical & Categorical Columns (Again)**

Sebelum saya melatih model, saya perlu memastikan semua fitur diproses dengan benar. Oleh karena itu, saya kembali mendefinisikan dengan jelas mana kolom numerik, mana kolom kategorikal yang bersifat ordinal (memiliki tingkatan), dan mana kolom kategorikal yang bersifat nominal (tidak ada tingkatan)

```python
# Kolom numerik (continuous)
numerical_cols_selected = ['subscription_length', 'seconds_of_use', 'frequency_of_use', 
                  'distinct_called_numbers', 'customer_value']

# Kolom ordinal (karena ada urutannya)
ordinal_cols_selected = ['charge_amount', 'age_group']

# Kolom kategorikal nominal (tanpa urutan)
categorical_cols_selected = ['complains', 'tariff_plan', 'status']

print(f"Kolom Numerik: {numerical_cols_selected}")
print(f"Kolom Ordinal: {ordinal_cols_selected}")
print(f"Kolom Kategorikal: {categorical_cols_selected}")
```

### **5.9. Column Transformer**

Selanjutnya, saya menyiapkan alat untuk pra-pemrosesan, yaitu `ColumnTransformer` yang saya beri nama `preprocessor`.
* Pada kolom numerik (`numerical_cols_selected`) akan diterapkan `StandardScaler` untuk mengubah skala nilainya (rata-rata 0, standar deviasi 1).
* Pada kolom kategorikal ordinal (`ordinal_categorical_cols` seperti `charge_amount`, `age_group`)akan diterapkan `OrdinalEncoder`.
* Pada kolom kategorikal nominal (`categorical_cols_selected` seperti `complains`, `tariff_plan` dll)akan diterapkan `OneHotEncoder`.

```python
preprocessor= ColumnTransformer(
    transformers=[
        ('scaler', StandardScaler(), numerical_cols_selected),
        ('ord_encode', OrdinalEncoder( handle_unknown='use_encoded_value', 
                                      unknown_value=-1), # Mengatasi unknown di test
                                      ordinal_cols_selected),
        ('cat_ohe', OneHotEncoder(handle_unknown='ignore', drop='first', sparse_output=False), 
                          categorical_cols_selected)
    ])
```

# **6. Model Definition**
---
Setelah data siap, tahap selanjutnya adalah pembuatan Pipeline untuk masing-masing algoritma  algoritma (model) yang akan digunakan untuk mempelajari pola dari data latih dan melakukan prediksi.

```python
# categorical_features for SMOTENC berdasarkan hasil sebelumnya
categorical_features_for_smotenc = [5, 6]  # charge_amount (idx 5) dan age_group (idx 6)

# Full pipeline
pipeline = ImbPipeline(steps=[
    ('preprocessing', preprocessor),
    ('smote', SMOTENC(categorical_features=categorical_features_for_smotenc, random_state=42)),
    ('classifier', RandomForestClassifier(random_state=42))
])
```

# **7. Model Training**
---
Pada tahap ini adalah untuk melatih model-model regresi yang telah didefinisikan pada Tahap 6. Proses training menggunakan data latih yang sudah melalui semua tahap preprocessing.

```python
# Fit model
pipeline.fit(X_train_selected, y_train)
```
